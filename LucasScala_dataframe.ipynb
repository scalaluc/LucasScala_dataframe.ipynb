```python
#1
import pandas as pd

times = ["Los Angeles Lakers", "Boston Celtics", "Golden State Warriors", "San Antonio Spurs", "Chicago Bulls", "Dallas Mavericks"]
ano = [1947, 1946, 1946, 1967, 1966, 1980]
titulos = [17, 17, 7, 5, 6, 1, 1]
conferencia = ["Oeste", "Leste", "Oeste", "Oeste", "Leste", "Oeste"]
idolo = ["Kobe Bryant", "Bill Russell", "Stephen Curry", "Tim Duncan", "Michael Jordan", "Dirk Nowitzki"]
colocacao = [7, 1 , 10, 14, 9, 5]
dados = list(zip(times, ano, titulos, conferencia, idolo, colocacao) )

df = pd.DataFrame(data = dados)
df.columns = ['Nome do time', 'Ano de Fundação','Titulos', 'Conferência', 'Idolo', 'Colocação']


print(df)
```

```python
#2
#Usei o chat gpt pra ver a maneira mais facil, tinha mais uma só q mais complicada\para ver o numero de linahs e colunas
numero_linhas, numero_colunas = df.shape

print("Número de linhas:", numero_linhas)
print("Número de colunas:", numero_colunas)
```

```python
#3 ver um item especifico com todas as suas especificações
linha_iloc = df.iloc[1]
print("\n", linha_iloc)
```

```python
#4 verificar se o data frame esta vazio
if df.empty:
    print("O DataFrame está vazio.")
else:
    print("O DataFrame não está vazio.")
```

```python
#5 auto explicativo
primeiros_registros = df.head()
print("Os 5 primeiros registros:\n", primeiros_registros)
```

```python
#6. Excluir um item
df.drop(1)
```

```python
#7 Adicionar uma novo item a lista


new_row = {'Nome do time': 'Miami Heat', 'Ano de Fundação': 1988, 'Titulos': 3, 'Conferência': 'Leste', 'Idolo': 'Dwyane Wade', 'Colocação': 8}

df.loc[6]=new_row
print(df)
```

```python
#8 transponha a coluna para linha

df_transposto = df.transpose()

print(df_transposto)
```

```python
#9 Mostre na tela as 2 primeiras colunas

df.iloc[:,:2]
```
